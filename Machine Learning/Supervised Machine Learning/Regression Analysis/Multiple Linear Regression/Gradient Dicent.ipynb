```python
import pandas as pd
import numpy as np
```

```python
data = {
    "size" : [2.104,1.416,1.534,0.852],
    "room" : [5,3,3,2],
    "floor" : [1,2,2,1],
    "house age" : [4.5,4.0,3.0,3.6],
    "price" : [4.60,2.32,3.15,1.78]
}
```

```python
dataset = pd.DataFrame(data)
```

```python
dataset
```

```python
x = np.array(dataset.iloc[:,:-1])
```

```python
x[:,0]
```

```python
y = np.array(dataset["price"])
```

```python
m = len(y)
```

```python
theta0 = 100
theta1 = 10
theta2 = 2
theta3 = -48
theta4 = 1
```

```python
def pred(x,theta0,theta1,theta2,theta3,theta4):
    return  theta0+(theta1*x[:,0])+(theta2*x[:,1])+(theta3*x[:,2])+(theta4*x[:,3])
```

```python
def MSE(y,y_pred):
    return (1/m) * np.sum((y - y_pred)**2)
```

```python
def d0(x,y,theta0,theta1,theta2,theta3,theta4):
    return (-2/m)*np.sum(y-pred(x,theta0,theta1,theta2,theta3,theta4))
```

```python
def d1(x,y,theta0,theta1,theta2,theta3,theta4):
    return (-2/m)*np.sum((y-pred(x,theta0,theta1,theta2,theta3,theta4))*x[:,0])
```

```python
def d2(x,y,theta0,theta1,theta2,theta3,theta4):
    return (-2/m)*np.sum((y-pred(x,theta0,theta1,theta2,theta3,theta4))*x[:,1])
```

```python
def d3(x,y,theta0,theta1,theta2,theta3,theta4):
    return (-2/m)*np.sum((y-pred(x,theta0,theta1,theta2,theta3,theta4))*x[:,2])
```

```python
def d4(x,y,theta0,theta1,theta2,theta3,theta4):
    return (-2/m)*np.sum((y-pred(x,theta0,theta1,theta2,theta3,theta4))*x[:,3])
```

```python
lr = 0.001
```

```python
max_itr = 1000000
```

```python
tar = 0.00001
```

```python
for i in range(max_itr):
    dl0 = d0(x,y,theta0,theta1,theta2,theta3,theta4)
    dl1 = d1(x,y,theta0,theta1,theta2,theta3,theta4)
    dl2 = d2(x,y,theta0,theta1,theta2,theta3,theta4)
    dl3 = d3(x,y,theta0,theta1,theta2,theta3,theta4)
    dl4 = d4(x,y,theta0,theta1,theta2,theta3,theta4)

    theta0 = theta0-(lr*dl0)
    theta1 = theta1-(lr*dl1)
    theta2 = theta2-(lr*dl2)
    theta3 = theta3-(lr*dl3)
    theta4 = theta4-(lr*dl4)

    pr = pred(x,theta0,theta1,theta2,theta3,theta4)
    mse = MSE(y,pr)

    if mse <= tar:
        print(i)
        break
```

```python
t0 , t1 ,t2 , t3 , t4 = theta0 , theta1 ,theta2 , theta3 , theta4
```

```python
pred = t0 + (t1*2.104) + (t2*5)  +(t3*1) +(t4*4.5)
```
