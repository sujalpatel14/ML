```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("multi_linear_regression_1000_rows.csv")
```

```python
dataset.shape
```

```python
dataset.isnull().sum()
```

```python
sns.pairplot(data=dataset)
plt.show()
```

```python
sns.heatmap(data=dataset.corr(),annot=True)
```

```python
x = dataset.iloc[:,:-1]
```

```python
y = dataset['Salary']
```

```python
from  sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,train_size=0.20,random_state=42)
```

```python
mlr = LinearRegression()
```

```python
mlr.fit(x_train,y_train)
```

```python
mlr.predict([[30,5]])
```

```python
mlr.score(x_test,y_test)*100
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,mlr.predict(x_test))
```
