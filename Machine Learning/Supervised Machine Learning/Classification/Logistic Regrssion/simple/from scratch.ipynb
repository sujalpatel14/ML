```python
import pandas as pd
import numpy as  np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("Social_Network_Ads.csv")
```

```python
dataset.head()
```

```python
dataset.drop(columns=["EstimatedSalary"],inplace=True)
```

```python
dataset.info()
```

```python
x =  dataset["Age"]
```

```python
x = (x-x.mean())/x.std()
```

```python
y = dataset["Purchased"]
```

```python
m = ( np.sum((x-x.mean())*(y-y.mean())) ) / np.sum(np.pow(x-x.mean(),2))
```

```python
b = y.mean() - (m*x.mean())
```

```python
z = (m*x+b)
```

```python
predict = 1 / (1+np.exp(-z))
```

```python
predict = (predict>=0.5).astype(int)
```

```python
plt.figure(figsize=(5,5))
sns.scatterplot(x='Age',y='Purchased',data=dataset)
sns.lineplot(x='Age',y=predict,data=dataset,color='r')
plt.show()
```
