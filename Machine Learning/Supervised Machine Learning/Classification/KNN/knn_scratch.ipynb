```python
import pandas as pd
import math
from sklearn.datasets import load_iris
import statistics as st
from sklearn.model_selection import train_test_split
```

```python
dataset = load_iris()
```

```python
dataset.feature_names
```

```python
x = dataset.data
```

```python
y = dataset.target
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
class KNN:
    def __init__(self,k):
        self.x_train = list()
        self.y_train = list()
        self.k = k
        
    def fit(self,x_train,y_train):
        self.x_train = x_train
        self.y_train = y_train

    def predict(self,new):
        distances = []
        for i in self.x_train:
            distance = 0
            for j in range(len(i)):
                distance += (i[j]-new[j])**2
            distances.append(math.sqrt(distance))

        pairs = []

        for i in range(len(distances)):
            pairs.append((distances[i],self.y_train[i]))

        pairs.sort(key=lambda x: x[0])

        k_nearest = pairs[:self.k]

        labels = [label for _,label in k_nearest]
        
        return st.mode(labels)

    def score(self,x_test,y_test):
        total_true = 0
        total_false = 0
        
        for i in range(len(x_test)):
            if y_test[i] == self.predict(x_test[i]):
                total_true += 1
            else :
                total_false += 1
                
        return total_true/len(y_test)
```

```python
kn = KNN(3)
```

```python
kn.fit(x_train,y_train)
```

```python
kn.predict([5.1, 3.5, 1.4, 0.2])
```

```python
kn.score(x_test,y_test)
```

```python
for k in range(1,100):
    current_score = 0
    kn = KNN(k)
    kn.fit(x_train,y_train)
    print(k,"  ",kn.score(x_train,y_train))
```
