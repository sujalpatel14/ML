```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("loan.csv")
```

```python
#first five row
dataset.head()
```

```python
#total row and column
dataset.shape
```

```python
#check all mising value in each row
dataset.isnull().sum()
```

```python
#total missing value in overall dataset
dataset.isnull().sum().sum()
```

```python
#total missing value in overall dataset form of percentage
(dataset.isnull().sum().sum()/(dataset.shape[0]*dataset.shape[1]))*100
```

```python
#total missing value in each row of dataset form of percentage
(dataset.isnull().sum()/dataset.shape[0])*100
```

```python
#total no of not null value in dataset
dataset.notnull().sum()
```

```python
#check missing value using graph for better understanding
sns.heatmap(dataset.isnull())
plt.show()
```
