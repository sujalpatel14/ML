# Encoding For Nominal Data

```python
import pandas as pd
```

```python
df = pd.DataFrame({"Name":["wscube","cow","cat","dog","black"]})
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
df["en_name"]=le.fit_transform(df["Name"])
```

# Encoding For Ordinal Data

```python
df = pd.DataFrame({"Size":['s','m','l','xl','s','m','l','s','s','l','xl',"m"]})
```

```python
df.head()
```

```python
ord_data = [["s","m","l","xl"]]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories=ord_data)
oe.fit(df[["Size"]])
```

```python
df["size_en"]=oe.transform(df[["Size"]])
```

df

Using Map Function

```python
ord_data1 = {"s":0,"m":1,"l":2,"xl":3}
```

```python
df["size_en_map"]=df["Size"].map(ord_data1)
```

```python
df
```

Perform On Real Dataset

```python
dataset = pd.read_csv("loan.csv")
```

```python
oen = OrdinalEncoder(categories=en_data_ord)
```
